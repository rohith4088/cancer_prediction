```python
conv = {'B':0,'M':1}
y = y.map(conv)
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
x_scaled = sc.fit_transform(x)
```

```python
x = pd.DataFrame(x_scaled,columns = x.columns)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.4,random_state = 42)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
import pickle 
scaling = pickle.dump(scaler,open("scaling.pkl",'wb'))

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold,cross_val_score
from sklearn.neighbors import KNeighborsClassifier as knn
k_values = np.arange(1,21)
mean_score = []
for k in k_values:
    clf1 = knn(n_neighbors = k)
    kfold = KFold(n_splits = 5,shuffle = True,random_state = 42)
    scores = cross_val_score(clf1,x,y,cv = kfold,scoring = 'accuracy')
    mean_score.append(np.mean(scores))
plt.plot(k_values,mean_score)
plt.xlabel("N_neighbors")
plt.ylabel("mean validation accuracy score")
plt.grid(True)
plt.show()
optimal_k = k_values[np.argmax(mean_score)]
print(f'the most optimal value of k is {optimal_k}')
```

```python
clf = knn(n_neighbors=7,algorithm='kd_tree')
```

```python
clf.fit(xtrain,ytrain)
```

```python
y_pred = clf.predict(xtest)
```

```python
y_pred.shape
```

```python
comp = pd.DataFrame(np.c_[ytest,y_pred],columns = ['oroginal','predicted'])
comp
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(ytest,y_pred))
```
